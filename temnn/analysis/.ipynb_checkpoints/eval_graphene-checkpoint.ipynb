```python
%matplotlib
%load_ext autoreload
%autoreload 2
import numpy as np
import numpy.ma as ma
import matplotlib.pyplot as plt
from glob import glob

from pyqstem.imaging import CTF
from pyqstem import PyQSTEM
from pyqstem.util import atoms_plot

import tensorflow as tf
from .net import net,mods
from net.dataset import DataEntry,DataSet

from ase.io import read

from ptm.template_matching import TemplateMatching
from ptm.peaks import find_local_peaks, refine_peaks
from ptm.markers import MarkerCollection, MarkerEditor, heatmap

from scipy.spatial.distance import cdist


from scipy.signal import savgol_filter
from IPython.display import clear_output
plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=17, serif='cm10')
```

```python
def load(data_dir):
    
    models=glob(data_dir+"model/model_*.cfg")
    waves=glob(data_dir+"wave/wave_*.npz")
    labels=glob(data_dir+"label/label_*.npy")
    entries=[DataEntry(model,wave,label) for model,wave,label in zip(models,waves,labels)]
    
    return DataSet(entries)

def label_edges(markers,limits):
    coords=markers.coords
    
    mask = np.zeros(coords.shape[0],dtype=bool)
    mask[coords[:,0]<limits[0]] = True
    mask[coords[:,0]>limits[1]] = True
    mask[coords[:,1]<limits[2]] = True
    mask[coords[:,1]>limits[3]] = True
    
    labels=markers.labels
    labels[mask]='edge'
    markers.set_labels(labels)
    return markers

def evaluate(truth,out,tol,neigh,thres,refine=0):
    #size=out.shape
    edge=4*tol

    peaks,values=find_local_peaks(out,neigh,global_thres=thres,return_values=True)
    if refine>0:
        peaks=refine_peaks(out,peaks,refine)
    
    peaks=MarkerCollection(peaks[:,1],peaks[:,0])
    peaks.set_property('probability',values)

    if peaks.num_markers > 0:
        closest=np.argmin(cdist(peaks.coords,truth.coords),axis=1)
        peaks.set_labels(truth.labels[closest])
        #peaks = label_edges(peaks,limits)
        
        peaks.set_property('fp',np.min(cdist(peaks.coords,truth.coords),axis=1)>tol)
        truth.set_property('fn',np.min(cdist(truth.coords,peaks.coords),axis=1)>tol)
    
    return truth,peaks

def precision_recall(truth,peaks,label):
    
    if peaks.num_markers>0:
        fn = np.sum(truth.get_property('fn',label))
        fp = np.sum(peaks.get_property('fp',label))
        tp = len(peaks.get_coords(label=label))-fp
        precision = tp/(tp+fp)
        recall = tp/(tp+fn)
    else:
        precision = 0
        recall = 0
    
    return precision, recall

def pr_curve(peaks,truth,label,N=40):
    
    
    prob=peaks.get_property('probability')
    
    precision = np.zeros(N)
    recall = np.zeros(N)
    for i,thres in enumerate(np.linspace(0,1,N)):
        
        new_peaks = peaks.copy()
        new_peaks.markers = [marker for marker,p in zip(new_peaks.markers,prob) if p>thres]
        
        p,r=precision_recall(truth,new_peaks,label)
        precision[i]=p
        recall[i]=r
    
    return precision,recall

def best_f_score(precision,recall):
    divisor=precision+recall
    divisor=ma.masked_array(divisor,mask=divisor==0)
    
    f_score=2*(precision*recall)/divisor
    return f_score.max(), precision[f_score.argmax()], recall[f_score.argmax()]

def create_truth(entry,edge=30):
    truth=entry._model.get_positions()/sampling
    truth=MarkerCollection(truth[:,1],truth[:,0])
    template_matching(truth,'hexagonal',strain=True,match_partial=False,scale=1.42,tol=.05)

    labels=np.array(['hexagonal']*truth.num_markers)
    labels[np.isnan(truth.get_property('rmsd'))]='other'

    truth.set_labels(labels)
    
    limits=[edge,size[0]-edge,edge,size[1]-edge]
    truth = label_edges(truth,limits)
    
    return truth
        
def harmonic_mean(values):
    return len(values)/np.sum([1/(np.array(v)+1e-12) for v in values],axis=0)
```

```python
model_path='models/model.ckpt'

batch_size=1
size=(1024,1024)
kernel_num=32
image_features=1
num_classes=1

tf.reset_default_graph()

x = tf.placeholder(tf.float32, shape=(batch_size,)+size+(image_features,))
y = FusionNet.graph(x,kernel_num,output_features=num_classes)

saver = tf.train.Saver()
sess=tf.Session()
sess.run(tf.global_variables_initializer())
saver.restore(sess, model_path)
```

```python
data_dir="../tem-neural-datasim/datasets/graphene-grains-md/"

data = load(data_dir)

sampling=244.8/2048
Cs=-11.5*10**4
defocus=-80/400000.*Cs+60
focal_spread=28
blur=0.2
doses=10**np.arange(1,5.25,.25)
mtf_param=[1,0,4.89683027e-01,2.34644273e+00]

images={}
truths={}

for i in range(16):
    entry=data.next_batch(1)[0]
    entry.load()

    ctf=CTF(defocus=defocus,Cs=Cs,focal_spread=focal_spread)
    
    images[i]={}
    for j,dose in enumerate(doses):
        
        entry.create_image(ctf,sampling,blur,dose,mtf_param)
        entry.local_normalize(30,60)
        entry.normalize()
        
        image,label=entry.as_tensors()
        images[i][j]=image
        
        entry.reset()
    
    truths[i]=create_truth(entry)
```

```python
f_scores={}

for i in images.keys():
    f_scores[i]={}
    truth=truths[i]
    
    for j,image in images[i].items():
        f_scores[i][j]={}
        
        #print(image.shape)
        
        out=sess.run(y, feed_dict={x: image})
        
        truth,peaks=evaluate(truth,out[0,:,:,0],4,6,0)
        
        for label in ['other','hexagonal']:
            precision,recall=pr_curve(peaks,truth,label)
            f_scores[i][j][label]={}
            
            f_scores[i][j][label]['f_score']=best_f_score(precision,recall)[0]
            f_scores[i][j][label]['optimal_precision']=best_f_score(precision,recall)[1]
            f_scores[i][j][label]['optimal_recall']=best_f_score(precision,recall)[2]
            
            f_scores[i][j][label]['precision']=precision
            f_scores[i][j][label]['recall']=recall
            f_scores[i][j][label]['fn'] = np.sum(truth.get_property('fn',label))
            f_scores[i][j][label]['fp'] = np.sum(peaks.get_property('fp',label))
        
        print(j,i)
        clear_output(wait=True)
```
